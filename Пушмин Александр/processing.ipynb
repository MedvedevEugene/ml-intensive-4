Преобразования данных с добавлением новых столбцов


```python
import pandas as pd

df = pd.read_csv('разметка комментраиев.csv')
df.dropna(how='all', inplace=True)
df.to_csv('данные.csv', index=False, encoding='utf-8', sep=',')
```

Инициализация нужных данных


```python
df = pd.read_csv('данные.csv')

df
```

Удаление пропусков после разметки данных

```python
df_clear = df.dropna(how='all')

df_clear
```

Удаление ненужных столбцов

```python

df = df_clear.drop(columns=['annotator', 'annotation_id', 'created_at', 'lead_time', 'updated_at'], axis=1)

df
```

Цикл для заполенения столбцов категориями

```python
categories = [
    'Вопрос решен',
    'Нравится качество выполнения заявки',
    'Нравится качество работы сотрудников',
    'Нравится скорость отработки заявок',
    'Понравилось выполнение заявки',
    'Работа не выполнена'
]


for cat in categories:
    df[cat] = df['taxonomy'].apply(
        lambda x: (
            int(
                cat in [
                    subitem[0]
                    for item in ast.literal_eval(x) if isinstance(item, dict) and 'taxonomy' in item
                    for subitem in item['taxonomy']
                ]
            ) if isinstance(x, str) and x.startswith('[') else 0
        ) if isinstance(x, str) else 0  
    )
```

Изменение индекса и удаление ненужного столбца

```python
df.reset_index(drop=True, inplace=True)
df.index += 1

df = df.drop('taxonomy', axis=1)
```

Удаление дубликатов комментариев

```python
df_cleaned = df.drop_duplicates(subset=["comment"]).reset_index(drop=True)
```

 Подготовка данных

```python
target_columns = [
    "Вопрос решен",
    "Нравится качество выполнения заявки",
    "Нравится качество работы сотрудников",
    "Нравится скорость отработки заявок",
    "Понравилось выполнение заявки",
    "Работа не выполнена"
]
df_model = df_cleaned[["comment"] + target_columns]

label_distribution = df_model[target_columns].sum().sort_values(ascending=False)
df_model.shape, label_distribution
```
